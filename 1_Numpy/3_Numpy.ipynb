## 7. Comparison

### all & any

```python
a = np.arange(8)
a
```

```python
a > 4
```

```python
np.all(a > 4), np.any(a > 4)
```

```python
x = np.array([1,3,0])
y = np.array([5,2,1])
x > y
```

```python
x == y
```

### logical operation

```python
a = np.array([1,3,0],float)
b = np.logical_and(a>0, a<3)    # 0 < a < 3
b
```

```python
c = np.logical_not(b)
c
```

```python
np.logical_or(b, c)
```

```python
np.logical_xor(b, c)
```

```python
np.where(a > 1, 10, 0)      # where(condition, True일때의 값, False일때의 값)
```

```python
np.where(np.arange(10,20) > 15)     # index 값을 반환
```

```python
# Infinite, Not a Number
a = np.array([np.Inf, np.NaN, 1, np.nan, np.inf, np.NAN])
a
```

```python
np.isnan(a)             # Is not a number?
```

```python
np.isfinite(a)          # Is finite number?
```

```python
np.isnan(a).sum()
```

### argmax, argmin

```python
a = np.array([1,2,4,5,8,78,23,3])
np.argmax(a), np.argmin(a)
```

```python
a = a.reshape(2,-1)
np.argmax(a), np.argmin(a)
```

```python
index = 5
row, col = index // 4, index % 4
row, col
```

```python
a[row, col]
```

```python
np.argmax(a, axis=1), np.argmin(a, axis=1)
```

## 8. Boolean & fancy index

### boolean index

```python
a > 4
```

```python
a[a>4]
```

### fancy index

```python
a = np.array([2,4,6,8])
b = np.array([1,3,2,0,0,1])
a[b]
```

```python
a.take(b)
```

```python
a = np.arange(1,13).reshape(3,4)
b = np.array([1,2,0,1])         # 행 인덱스
c = np.array([2,1,3,0])         # 열 인덱스
a[b, c]
```
